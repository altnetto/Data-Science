<h3>Aula 003</h3>

One-hot encoding scheme

- Criar uma matriz de atributos ao invés de assumir números (ex: matriz RGB)

Ao invés de primeiro = 1, segundo = 2 e terceiro = 3,

Primeiro    |   Segundo     |   Terceiro
------------|---------------|---------------
    1       |      0        |       0
    0       |      1        |       0
    0       |      0        |       1

O que pode formar os objetos:

Primeiro = (1,0,0), 
Segundo = (0,1,0), 
Terceiro = (0,0,1)

<h3>Tipos de amostragem</h3>

- Simples
    - Sem reposição (dados são coletados aleatoriamente, sem considerar se já foi coletado)
    - Com reposição (dados coletados com cache)

- Estratificada
    - Classes com elementos diferentes (ex: Homens e Mulheres)
    - Tenta-se manter a proporção macro na amostra, para evitar grandes variâncias de dados
    
- Progressiva
    - Considera-se o aumento na melhoria dos dados, até que não se verifique mais mudanças dentro do escopo da qualidade de dados

<h3>Balanceamento</h3>

- Dependendo do problema e do local de coleta de dados, os resultados observados estarão desbalanceados, ou seja, não reprentam fidedigmamente a realidade

    - ex: 80% das pessoas que vão ao hospital estão doentes

    - Nesses casos, deve-se efetuar o Balanceamento

-------

<h3>Técnicas de Balanceamento</h3>

- Acréscimo: adicionar dados de uma determinada classe para diminuir o desbalanceamento
    - Pode acarretar em overfitting

- Remoção: remove dados de uma determinada classe, para diminuir o desbalanceamento
    - Pode gerar underfitting

- Custos (pesos) de classificação diferentes (ponderamento):
    - Usar lógica de Fuzzy para o ponderamento?

------

<h3>Limpeza dos Dados</h3>

- Ruídos
    - ex: Idade com valores negativos

- Inconsistências
    - ex: pessoa adulta pesando 10kg

- Redundâncias
    - ex: objetos/atributos com mesmos valores
        - Valores podem ser obtidos a partir de outros atributos (ex: valor unit. e valor total)
        - Atributos com a mesma info preditiva (ex: Idade e data de nascimento)
        - Pode tornar mais lento o processo de indução

- Dados incompletos
    - ex: ausência de salário para crianças
        - Eliminar os objetos com valores ausentes
        - Preenchimento manual
        - Empregar método/heurística para definir os valores automaticamente
        - Empregar algoritmos de AM que lidam internamente com valores ausentes

<h3>Transformação de Atributos</h3>

- Normalização: min-max-scaling
    - $Zi = (Xi-min(X))/(max(X)-min(X))$

- Padronização: Z-score normalization
    - $Zi = (Xi - \mu)/\sigma$

- Softmax function

- Tradução
    - ex: converter °F para °C
    - converter coordenadas geográficas por código postal
    - data de nascimento -> idade

<h3>Análise dos Componentes Principais</h3>

- Coeficiente de correlação de Pearson
    - $p = -1$ -> Anti-correlacionado
    - $p = 1$ -> correlacionado
    - $p = 0$ -> não correlacionado

- Basicamente, podemos transformar a nossa dispersão de modo que a nossa dispersão principal assuma o eixo Xn (abscissas) e a dispersão de variância seja estabelecida no eixo Yn (ordenadas), de modo a termos uma nova dispersão em apenas 1 dimensão

    - $A = Q * A * Q^{-1}$

        - Q: Matriz de autovetores de A
        - A: Matriz de autovalores de A
        - $Q^{-1}$: Matriz inversa de Q
            - $Q^{-1}$ só existe se os autovetores são linearmente independentes
