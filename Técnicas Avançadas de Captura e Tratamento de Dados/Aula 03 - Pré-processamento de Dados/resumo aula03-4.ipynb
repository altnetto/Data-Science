### Transformação de Atributos ###

Em vários momentos, podemos nos deparar com atributos que apresentam uma distribuição extremamente compacta, o que dificulta muito sua visualização e compreensão. Desse modo, é interessante possuir ferramentas para suavizar essa distribuição, permitindo que seja mais fácil sua análise.


Assim sendo, algumas técnicas de transformação de atributos são:

- Tomar o logaritmo do atributo

- Tomar a tangente do atributo

- Tomar a raíz quadrada do atributo

- dentre outras

### Assimetria (obliquidade ou skewness) ###

Em casos onde a distribuição dos dados se aproxima de uma normal, mas não é regular (distorcida com relação à normal), temos métodos de tratamento dos dados para que possamos compensar a assimetria e, então, podermos utilizar as propriedades da normal

- 0: normalmente distribuída (ou perfeitamente simétrica)
- \> 0: maior concentração na cauda esquerda da distribuição (negativa)
- < 0: maior concentração na cauda direita da distribuição (positiva)
